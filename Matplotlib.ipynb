```python
import numpy as np
x=np.array((1,2,3,4))
y=np.array((3,6,9,12))
# plt.plot(x,y)
#plt.plot(x,y,marker='o')
plt.plot(x,y,marker='D',ms=20,mec='r',mfc='k',ls=':',color='g',lw=10)
plt.show()
```

```python
import numpy as np
x=np.array((1,2,3,4))
y1=np.array((3,6,9,12))
y2=np.array((4,8,12,16))
plt.plot(x,y,marker='D',ms=20,mec='r',mfc='k',ls=':',color='g',lw=10)
plt.xlabel('B6',family='serif',color='blue',size=20)
plt.ylabel('VHA')
plt.title('vishal10',loc='left')
plt.grid(axis='x')
plt.show()
```

```python
import numpy as np
x=np.array((1,2,3,4))
y=np.array((3,6,9,12))
plt.plot(x,y1,marker='D',ms=20,mec='r',mfc='k',ls=':',color='g',lw=10,label='student1')
plt.plot(x,y2,label='student1')
plt.xlabel('B6',family='serif',color='blue',size=20)
plt.ylabel('VHA')
plt.title('vishal10',loc='left')
plt.grid(axis='x')
plt.legend(loc='upper right')
plt.show()
```

```python
##Bar Chart
import numpy as np
x=np.array(('PS','FSD','Python','DE'))
y=np.array(('24','9','10','7'))
#plt.bar(x,y,color='red')
plt.bar(x,y,color=['r','b','g','y'],width=0.2)
plt.show()
```

```python
##horizontal bar
##Bar Chart
import numpy as np
x=np.array(('PS','FSD','Python','DE'))
y=np.array(('24','9','10','7'))
#plt.bar(x,y,color='red')
plt.barh(x,y,color=['r','b','g','y'],height=0.2)
plt.show()
```
