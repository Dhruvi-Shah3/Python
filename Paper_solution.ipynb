```python
import numpy as np
import matplotlib.pyplot as plt
from abc import ABC,abstractmethod
class Data(ABC):
    Array=np.array([[64,13,100,46,69,85],
                   [54,43,7,34,17,19],
                   [73,73,90,77,53,14],
                   [74,41,41,83,74,95],
                   [48,31,74,49,67,99]])
    @abstractmethod
    def abcmethod(self):
        pass
class match(Data):
        def abcmethod(self):
            pass
        def match_highscore(self,match_number):
            return np.max(Data.Array[match_number-1])
        def player_lowscore(self,player_number):
            return np.min(Data.Array[:,player_number-1])
        def match_total(self):
            match_total=np.sum(Data.Array,axis=1)
            Data.Array=np.concatenate((Data.Array,match_total.reshape(5,-1)),axis=1)
            return Data.Array
        def sorted_matchtotal_hightolow(self):
            return Data.Array[np.argsort(Data.Array[:,-1])]
        def player_totalscore(self):
            return np.sum(Data.Array,axis=0)
        def binary_array(self):
            return np.where(Data.Array>50,'1','2')
        def lineplot(self,player1,player2):
            x=[1,2,3,4,5]
            plt.plot(x,Data.Array[:,player1-1])
            plt.plot(x,Data.Array[:,player2-1])
            plt.show()
        def pieplot(self,match_number):
            x=[1,2,3,4,5,6]
            plt.pie(Data.Array[match_number-1,:-1],labels=x)
            plt.show()
```

```python
obj=match()
print(obj.match_highscore(3))
print(obj.player_lowscore(2))
print(obj.match_total())
print(obj.sorted_matchtotal_hightolow())
print(obj.player_totalscore()) 
print(obj.binary_array()) 
print(obj.lineplot(3,4)) 
print(obj.pieplot(3)) 


                 
```
